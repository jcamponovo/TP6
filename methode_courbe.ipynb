## Afficher une courbe <a href="#Titre">(Sommaire)</a>

<b>1°)</b> Importer (charger en mémoire) la bibliothèque <b>matplotlib.pyplot</b> sous le nom <b>plt</b> ainsi que les fonctions et valeurs nécessaires.

```python
import matplotlib.pyplot as plt
from math import cos,pi
```

<b>2°)</b> Construire les listes de valeurs de x et de y. Elles doivent avoir le <b>même nombre de valeurs.</b>

```python
x = []
y = []
f = 4.
for i in range(100):
    x.append(i/100)
    y.append(3.5*cos(2*pi*f*x[i]))
```

<b>3°)</b> Construire le graphique y = f(x) puis l'afficher avec <b>plt.plot</b> et <b>plt.show</b>

```python
plt.figure("Tracer un graphique")
plt.plot(x,y)
plt.show()
```

## Mettre plusieurs graphiques sur une même figure <a href="#Titre">(Sommaire)</a>

<p>On utilise le mot clé <b>plt.subplot2grid</b> suivi de deux doublets de valeurs:</p><p>Le premier donne la taille de la grille (1,2) signifie 1 ligne et 2 colonnes.</p>
<p>Le deuxième donne le position du graphique. (0,1) signifie première ligne et deuxième colonne.</p>

```python
plt.figure("2 graphiques avec une grille 1x2 sur une même figure")
plt.subplot2grid((1,2),(0,0))
plt.plot(x1,y1,label="gauche")
plt.legend()
plt.subplot2grid((1,2),(0,1))
plt.plot(x1,y2,label="droite")
plt.legend()
plt.show()
```

<p>Ici (2,1) signifie 2 lignes et 1 colonne, (0,0) et (1,0) représentent respectivement le graphique du haut et le graphique du bas</p>

```python
plt.figure("2 graphiques avec une grille 2x1 sur une même figure")
plt.subplot2grid((2,1),(0,0))
plt.plot(x1,y1,label="haut")
plt.legend()
plt.subplot2grid((2,1),(1,0))
plt.plot(x1,y2,label="bas")
plt.legend()
plt.show()
```

<p>On utilise ici une grille 2x3, la position de chaque graphique est donné en légende</p>

```python
plt.figure("6 graphiques avec une grille 2x3 sur une même figure")
for i in range(3):
    for j in range(2):
        plt.subplot2grid((2,3),(j,i))
        plt.plot(x1,y1,label="Position ({},{})".format(j,i))
        plt.legend()
plt.show()
```

On peut aussi laisser des espaces vides dans la grille et étaler des graphiques sur plus d'un zone en utilisant les options facultatives <b>rowspan</b> et <b>colspan</b>.

```python
plt.figure("Utilisation avancée d'une grille")
plt.subplot2grid((2,4),(0,0))
plt.plot(x1,y1,label="Position (0,0) taille 1x1")
plt.legend()

plt.subplot2grid((2,4),(0,2),rowspan=2,colspan=2)
plt.plot(x1,y1,label="Position (0,2) taille 2x2")
plt.legend()

plt.subplot2grid((2,4),(1,0),colspan=2)
plt.plot(x1,y1,label="Position (1,0) taille 1x2")
plt.legend()

plt.show()
```

## Exemple complet <a href="#Titre">(Sommaire)</a>

```python
#on peut modifier A1, T1 et phi1 pour voir l'effet
A1 = 1.
T1 = 0.002
phi1 = 0.

A2 = 2.
T2 = 0.002
phi2 = 0.

t = []
Y1 = []
Y2 = []
dt = max(T1,T2)/10000

for i in range(30000): #on construit les listes de données
    t.append(i*dt)
    Y1.append(A1 * cos(2*pi*t[-1]/T1 + phi1))
    Y2.append(A2 * cos(2*pi*t[-1]/T2 + phi2))

plt.figure("Exemple complet")
    
plt.plot(t,Y1,"red",label="courbe 1") #on trace les deux courbes
plt.plot(t,Y2,"green",label="courbe 2")

ax = plt.gca() #ce bloc de 7 lignes permet d'avoir des axes qui se croisent en (0,0)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.xlabel("t (s)") #Ajoute une légende sur chaque axe
plt.ylabel("y (m)")

plt.legend() #affiche la légende du graphique

plt.grid() #affiche une grille

plt.title("Représenter deux sinusoides") #ajoute un titre à la figure

plt.show() #affiche le graphique
```

## Exemple d'animation <a href="#Titre">(Sommaire)</a>

```python
import matplotlib.animation as animation

fig = plt.figure("Exemple d'animation")
ax1 = plt.gca()
fig.show()

#Paramètres de la simulation
c = 340.
T = 1/1000.
dt = T/20.
xmax=3.
ig=0
ir=0

x3=[i*xmax/300. for i in range(300)]

#Fonction à représenter
def f(x,t,c,T):
    return cos(2*pi*(t/T-x/(c*T)))
   
#Fonction appelée lors de l'animation avec une valeur de i qui augmente de 1 à chaque fois    
def anime(i):
    global c,T,dt,xmax,ig,ir,x3    
    
    #Construire la liste des y pour le nouveau temps (i*dt)
    y=[]
    for j in range(300):
        y.append(f(x3[j],i*dt,c,T))
        
    #Construire la nouvelle courbe avec la légende correspondante
    ax1.clear()
    ax1.plot(x3,y,"blue",label="t = {:.3f}s".format(i*dt))
    
    #Si le point rouge sort de l'écran, on change de points 
    if c*(i-ir)*dt>xmax:
        ir=i
        ig=i
        
    #Point rouge
    ax1.plot([c*(i-ir)*dt],[f(c*(i-ir)*dt,i*dt,c,T)],"o",color="red")
    
    #point vert s'il est dans l'écran
    if c*(i-ig)*dt+c*T<xmax:
        ax1.plot([c*(i-ig)*dt+c*T],[f(c*(i-ig)*dt+c*T,i*dt,c,T)],"o",color="green")
        
    #Afficher la légende en haut à gauche de la figure
    ax1.legend(loc='upper right',framealpha=1.)

#Animation avec une image toutes les 50 ms
simulation=animation.FuncAnimation(fig,anime,interval=50)
```
